```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
```

```python
df=pd.read_csv('/city_sustainability_geospatial.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
pio.templates
```

```python
pio.templates.default='plotly'
```

```python
px.scatter(
    df,
    x='co2_tons_per_capita',
    y='air_quality_index',
    title='Co2 vs Air Quality Index',
    color='continent',
    size='population_millions'
)
```

```python
px.histogram(
    df,
    x='air_quality_index',
    title='Distribution of Air Quality Index',
    color='continent',
    marginal='box'
)
```

```python
px.box(
    df,
    y='population_millions',
    x='continent',
    title='populaion vs continent',
    color='continent',
    points='all'

)
```

```python
top_cities=df.nlargest(10,'population_millions').sort_values('population_millions',ascending=False)
```

```python
px.bar(
    top_cities,
    y='city',
    x='population_millions',
    title='Top 10 cities by population',
    color='continent'
)
```

```python
px.scatter_geo(
    df,
    lat='latitude',
    lon='longitude',
    size='population_millions',
    hover_data={
        'city':True,
        'population_millions':True,
        'continent':True,
        'co2_tons_per_capita':True,
        'air_quality_index':True,
        'renewable_energy_share':True,
        'longitude':False,
        'latitude':False
    },
    title='City Sustainability',
    color='continent',
    projection='natural earth'

)
```

```python
px.scatter_geo(
    df,
    lat='latitude',
    lon='longitude',
    size='population_millions',
    hover_data={
        'city':True,
        'population_millions':True,
        'continent':True,
        'co2_tons_per_capita':True,
        'air_quality_index':True,
        'renewable_energy_share':True,
        'longitude':False,
        'latitude':False
    },
    title='City Sustainability',
    color='air_quality_index',                                 #when we use color based on continuous data ,the result will be in a heat map
    projection='natural earth'

)
```

```python
country_status=df.groupby(['country','iso_alpha3','continent']).agg({
    'population_millions':'sum',
    'co2_tons_per_capita':'mean',
    'air_quality_index':'mean',
    'renewable_energy_share':'mean'
    }).reset_index()
```

```python
px.choropleth(
    country_status,
    locations='iso_alpha3',
    color='co2_tons_per_capita',
    hover_name='country',
    hover_data={
        'population_millions':True,
        'continent':True,
        'co2_tons_per_capita':True,
        'air_quality_index':True,
        'renewable_energy_share':True
    },
    title='CO2 Emissions by Country',
    projection='natural earth'
)
```

```python
px.scatter_mapbox(
    df,
    lat='latitude',
    lon='longitude',
    size='population_millions',
    color='air_quality_index',
    mapbox_style='open-street-map',
    hover_data={
        'city':True,
        'population_millions':True,
        'continent':True,
        'co2_tons_per_capita':True,
        'air_quality_index':True,
        'renewable_energy_share':True,
        'longitude':False,
        'latitude':False
    },
    color_continuous_scale='bupu',
    zoom=1,
    title='City Sustainability'
)
```
