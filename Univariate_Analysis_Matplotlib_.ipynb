```python
import pandas as pd
```

```python
df=pd.read_csv('/content/sales_dataset (1).csv')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,8))
```

```python
plt.hist(df['revenue'])
```

```python
plt.hist(df['revenue'])
plt.title('Revenue Distribution')     #if i wanna put title upon the figure
plt.xlabel('Revenue')                 #if i rename x axis
plt.ylabel('Frequency')               #if i rename y axis
```

```python
plt.hist(df['revenue'],bins=50)
```

```python
plt.figure()
plt.boxplot(df['revenue'])
plt.title('Revenue box plot')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.figure()
plt.violinplot(df['revenue'],showmeans=True,showmedians=True)
plt.title('Revenue violin plot')
plt.ylabel('Revenue')
plt.show()
```

```python
df.info()
```

```python
category_counts=df['product_category'].value_counts().reset_index()
category_counts
```

```python
plt.figure()
plt.bar(category_counts['product_category'],category_counts['count'])
plt.title('Revenue by product category')
plt.xlabel('product_category')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.hist(df['revenue'], bins=30, label='Revenue', alpha=0.5)
plt.hist(df['profit'], bins=30, label='Profit', alpha=0.5)
plt.legend()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].hist(df['revenue'], bins=30, label='Revenue', alpha=0.5)
axes[0].set_title('Revenue Distribution')
axes[0].set_xlabel('Revenue')
axes[0].set_ylabel('Frequency')
axes[1].hist(df['profit'], bins=30, label='Profit', alpha=0.5)
axes[1].set_title('Profit Distribution')
axes[1].set_xlabel('Profit')
axes[1].set_ylabel('Frequency')
plt.show()
```
