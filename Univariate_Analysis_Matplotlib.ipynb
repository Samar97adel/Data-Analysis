```python
import pandas as pd
```

```python
df = pd.read_csv('/content/sales_dataset (1).csv')
```

```python
df
```

```python
df.info()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['revenue'], bins=30, label='Revenue', alpha=0.5)
plt.hist(df['profit'], bins=30, label='Profit', alpha=0.5)
plt.legend()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].hist(df['revenue'], bins=30, label='Revenue', alpha=0.5)
axes[0].set_title('Revenue Distribution')
axes[0].set_xlabel('Revenue')
axes[0].set_ylabel('Frequency')
axes[1].hist(df['profit'], bins=30, label='Profit', alpha=0.5)
axes[1].set_title('Profit Distribution')
axes[1].set_xlabel('Profit')
axes[1].set_ylabel('Frequency')
plt.show()
```

```python
plt.figure()
counts, bins, patches = plt.hist(df['revenue'], bins='auto')
plt.title('Revenue Distribution')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.show()
```

```python
bins
```

```python
plt.figure()
plt.boxplot(df['revenue'])
plt.title('Revenue Box Plot')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.figure()
plt.violinplot(df['revenue'], showmeans=True, showmedians=True)
plt.title('Revenue Violin Plot')
plt.ylabel('Revenue')
plt.show()
```

```python
category_counts = df['product_category'].value_counts().reset_index()
category_counts
```

```python
plt.figure()
plt.bar(category_counts['product_category'], category_counts['count'])
plt.title('Revenue by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
pm_counts = df['payment_method'].value_counts().reset_index()
pm_counts
```

```python

plt.figure()
plt.pie(pm_counts['count'], labels=pm_counts['payment_method'], autopct='%1.1f%%')
plt.title('Payment method share')
plt.show()
```

```python
df.corr(numeric_only=True)
```

```python
plt.figure()
plt.scatter(df['unit_price'], df['revenue'], alpha=0.6)
plt.title('Unit Price vs Revenue')
plt.xlabel('Unit Price')
plt.ylabel('Revenue')
plt.show()
```

```python
monthly_rev = df.set_index('order_date').resample('M')['revenue'].sum()

plt.figure(figsize=(20, 8))
plt.plot(monthly_rev.index, monthly_rev.values)
plt.title('Revenue Over Time')
plt.xlabel('Order Date')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
plt.bar(df['segment'], df['revenue'])
plt.title('Revenue by Segment')
plt.xlabel('Segment')
plt.ylabel('Revenue')

plt.show()
```

```python
groups = []

for cat in df['product_category'].unique():
  groups.append(df.loc[df['product_category'] == cat, 'profit'])


plt.figure()
plt.boxplot(groups, labels=df['product_category'].unique())
plt.title('Profit by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Profit')
plt.show()
```

```python
groups[0]
```

```python
groups = []

for cat in df['segment'].unique():
  groups.append(df.loc[df['segment'] == cat, 'revenue'])


plt.figure()
plt.boxplot(groups, labels=df['segment'].unique())
plt.title('Revenue by Segment')
plt.xlabel('Segment')
plt.ylabel('Revenue')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['unit_price'], df['quantity'], df['profit'], c='r', marker='x', alpha=0.9)

ax.set_title('Unit Price vs Quantity vs Profit')
ax.set_xlabel('Unit Price')
ax.set_ylabel('Quantity')
ax.set_zlabel('Profit')
plt.show()
```
