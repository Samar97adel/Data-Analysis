#Categorical Data Cleaning

```python
df_clean=df_nodup.copy()
```

```python
df_clean.info()
```

```python
import numpy as np
```

```python
df_clean=df_clean['City'].astype(str).str.strip().str.title().str.replace(',','').replace('Nan',np.nan)
```

```python
df_clean['Gender']=df_clean['Gender'].astype(str).str.strip().str.title().str.replace(',','').replace('Nan',np.nan)
```

```python
df_clean['City'].value_counts(dropna=False).sort_values(ascending=False)
```

```python
df_clean['Gender'].value_counts(dropna=False).sort_values(ascending=False)
```

```python
df_clean['Age']=pd.to_numeric(df_clean['Age'].replace({'twenty five':25}),errors='coerce')
```

```python
df_clean['MonthlyIncome']=pd.to_numeric(df_clean['MonthlyIncome'].replace('unknown',np.nan),errors='coerce')
```

```python
df_clean.info()
```

```python
df_clean['Age'].sample(10)  #'twenty five' is become '25'
```

```python
df_clean['Age'].value_counts().sort_values(ascending=False)    #To know the age group is most common (18 years old)
```

#outliers check
#outlier scanning by IQR

```python
df_clean.info()
```

```python
s1=df_clean['Age'].dropna()
```

```python
q1 =s1.quantile(0.25)
q3=s1.quantile(0.75)
iqr=q3-q1
lower=q1-1.5*iqr
upper=q3+1.5*iqr
print(f'Q1:(q1)')
print(f'Q3:(q3)')
print(f'IQR:(iqr)')
print(f'Lower bound:(lower)')
print(f'Upper bound:(upper)')
```

```python
df_clean[df_clean['Age']<lower]
```

```python
 df_clean[df_clean['Age']>upper]
```

#outliers of Monthly income

```python
s2=df_clean['MonthlyIncome'].dropna()
```

```python
q1 =s2.quantile(0.25)
q3 =s2.quantile(0.75)
iqr= q3-q1
lower=q1-1.5*iqr
upper=q3+1.5*iqr
print(f'Q1:(q1)')
print(f'Q3:(q3)')
print(f'IQR:(iqr)')
print(f'Lower bound:(lower)')
print(f'Upper bound:(upper)')
```

```python
df_clean[df_clean['MonthlyIncome']<lower]
```

```python
df_clean[df_clean['MonthlyIncome']>upper]
```
